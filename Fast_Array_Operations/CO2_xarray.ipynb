# Analyzing CO2 Concentration with Pandas and Xarray

### Analyzing CO2 concentration with Pandas and then with xarray

```python
# Load CO2 data file
import pandas as pd
df = pd.read_csv('monthly_co2.csv', index_col=[0, 1])
df.head()
```

```python
# Plotting the increase in CO2 
import matplotlib.pyplot as plt 

plt.plot(df.co2.values)
```

```python
# Plottig the increase in CO2 by taking mean of every year (removes zig zag)
by_year = df.groupby('year').mean().co2
plt.plot(by_year)
```

```python
# Plottig the change in CO2 of each month by taking mean.
by_month = df.groupby('month').mean().co2
plt.plot(by_month)
```
