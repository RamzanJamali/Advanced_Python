# Analyzing CO2 Concentration with Pandas and Xarray

### Analyzing CO2 concentration with Pandas and then with xarray

```python
# Load CO2 data file
import pandas as pd
df = pd.read_csv('monthly_co2.csv', index_col=[0, 1])
df.head()
```

```python
# Plotting the increase in CO2 
import matplotlib.pyplot as plt 

plt.plot(df.co2.values)
```

```python
# Plottig the increase in CO2 by taking mean of every year (removes zig zag)
by_year = df.groupby('year').mean().co2
plt.plot(by_year)
```

```python
# Plottig the change in CO2 of each month by taking mean.
by_month = df.groupby('month').mean().co2
plt.plot(by_month)
```

## Analyzing with xarray

```python
# Loading data from panda's dataframe to xarray's dataset
import xarray as xr

ds = xr.Dataset.from_dataframe(df)
```

```python
# Showing data in database style
ds
```

```python
# Slicing data like numpy
ds['co2'][:10, :5]
```

```python
# Selecting data like pandas
ds['co2'].sel(year=1960)
```

### Improved Performance

```python
# Taking mean of months
ds.mean(dim='month')
```

```python
%timeit df.groupby('year').mean().co2
```

```python
%timeit ds.mean(dim='month').co2.values
```

### Plotting with xarray

```python
import numpy as np
measures = np.load('2d_measure.npy')
measures.shape
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10,3))

for i in range(3):
    c = ax[i].imshow(measures[:, :, i], origin='lower')
    plt.colorbar(c, ax=ax[i])
    
plt.show()
```

```python
da = xr.DataArray(measures, dims=['x', 'y', 'time'])
```

```python
da.plot.imshow(x='x', y='y', col='time', robust=True)
```
